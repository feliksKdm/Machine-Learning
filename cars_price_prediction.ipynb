# Car price prediction via mashina.kg


```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
df = pd.DataFrame()


for i in range(1, 28):


    print(i)
    req = requests.get(f"https://www.mashina.kg/search/subaru/?currency=2&price_from=&price_to=&page={i}")
    html = req.text
    soup = BeautifulSoup(html)
    divs_all = soup.find_all('div', class_ = 'list-item list-label')




    car_model = []
    year_of_production = []
    engine_volume = []
    mileage = []
    fuel = []
    gearbox_mechanism = []
    steering_wheel_location = []
    color = []
    body_type = []
    location = []
    price_in_USD = []




    for j in divs_all:




        car_model1 = j.find('h2', class_ = 'name').text.strip()
        year1 = j.find('p', class_ = 'year-miles').text.split()[0]
        engine1 = j.find('p', class_ = 'year-miles').text.split()[2]
        mileage1 = j.find('p', class_ = "volume").text.split()
        mileage2 = ''
        if len(mileage1) > 2:
            for i in range(2, len(mileage1) - 1):
                mileage2 += mileage1[i]
        else:
            mileage2 = 0
        fuel1 = j.find('p', class_ = 'body-type').text.strip().split()[-1]
        gearbox1 = j.find('p', class_ = 'year-miles').text.split()[-1]
        steering1 = j.find('p', class_ = "volume").text.split()[1].replace(',', '')
        color1 = j.find('i')['title']
        btype1 = j.find('p', class_ = 'body-type').text.strip().split()[0].replace(',', '')
        location1 = j.find('p', class_ = 'city').text.split()[0]
        price1 = j.find('div', class_ = 'block price').find('strong').text.split()
        price2 = ''
        for i in range(1, len(price1)):
            price2 += price1[i]




        car_model.append(car_model1)
        year_of_production.append(float(year1))
        engine_volume.append(float(engine1))
        mileage.append(float(mileage2))
        fuel.append(fuel1)
        gearbox_mechanism.append(gearbox1)
        steering_wheel_location.append(steering1)
        color.append(color1)
        body_type.append(btype1)
        location.append(location1)
        price_in_USD.append(float(price2))




    df1 = pd.DataFrame({'Car model' : car_model,
                       'Year of production' : year_of_production,
                       'Engine volume' : engine_volume,
                       'Mileage' : mileage,
                       'Fuel' : fuel,
                       'Gearbox mechanism' : gearbox_mechanism,
                       'Steering wheel location' : steering_wheel_location,
                       'Color' : color,
                       'Body type' : body_type,
                       'Location' : location,
                       'Price in USD' : price_in_USD})




    df = pd.concat([df, df1])
```

```python
df['Steering wheel location'] = pd.get_dummies(df['Steering wheel location'])
```

```python
df['Fuel'] = pd.get_dummies(df['Fuel'])
```
